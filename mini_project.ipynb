```python
defaultString = 'NAN'
mh_subset['work_interfere'] = mh_subset['work_interfere'].replace([defaultString], 'Don\'t know' )
print(mh_subset['work_interfere'].unique())
```

```python
# Assign default values for each data type
defaultInt = 0
defaultString = 'NaN'

# Create lists by data tpe
intFeatures = ['Age']
stringFeatures = ['Gender', 'Country','family_history', 'treatment', 'work_interfere'
                 , 'tech_company', 'anonymity', 'leave', 'mental_health_consequence',
                 'phys_health_consequence', 'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview',
                 'mental_vs_physical', 'obs_consequence', 'benefits', 'care_options', 'wellness_program',
                 'seek_help']

# Clean the NaN's
for feature in mh_subset:
    if feature in intFeatures:
        mh_subset[feature] = mh_subset[feature].fillna(defaultInt)
    elif feature in stringFeatures:
        mh_subset[feature] = mh_subset[feature].fillna(defaultString)
    else:
        print('Error: Feature %s not recognized.' % feature)
mh_subset.head(30)
```

```python
defaultString = 'NaN'
# replacing the NAN with No
# mh_subset['self_employed'] = mh_subset['self_employed'].replace([defaultString], 'No')
# print(mh_subset['self_employed'].unique())

# replacing the NAN with dont know
mh_subset['work_interfere'] = mh_subset['work_interfere'].replace([defaultString], 'Don\'t know' )
print(mh_subset['work_interfere'].unique())
```

DATA CLEANING

```python
#clean 'Gender'
#Slower case all columm's elements
gender = mh_subset['Gender'].str.lower()
#print(gender)

#Select unique elements
gender = mh_subset['Gender'].unique()

#Made gender groups
male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "Cis Male", "cis male"]
trans_str = ["trans-female", "something kinda male?", "queer/she/they", "non-binary","nah", "all", "enby", "fluid", "genderqueer", "androgyne", "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", "female (trans)", "queer", "ostensibly male, unsure what that really means"]
female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]

for (row, col) in mh.iterrows():

    if str.lower(col.Gender) in male_str:
        mh_subset['Gender'].replace(to_replace=col.Gender, value='male', inplace=True)

    if str.lower(col.Gender) in female_str:
        mh_subset['Gender'].replace(to_replace=col.Gender, value='female', inplace=True)

    if str.lower(col.Gender) in trans_str:
        mh_subset['Gender'].replace(to_replace=col.Gender, value='trans', inplace=True)

#Get rid of irrelevant details
stk_list = ['A little about you', 'p']
mh = mh[~mh['Gender'].isin(stk_list)]

print(mh_subset['Gender'].unique())
mh_subset
```

```python
#correlation 1
#countries and attitude towards mental health illnesses
north_america = ['United States','Canada','Mexico','Bahamas,The','Costa Rica']
south_america = ['Brazil','Colombia','Uruguay']
europe = ['United Kingdom','Spain','Bosnia and Herzegovina','Hungary','Romania','Georgia','Czech Republic','Norway','Slovenia','Germany','Ireland','Netherlands','France','Latvia','Poland','Switzerland','Sweden','Italy','Belgium','Bulgaria','Austria','Finland','Denmark','Greece','Croatia','Portugal','Moldova']
africa = ['South Africa','Nigeria','Zimbabwe']
asia = ['Israel','India','Singapore','Russia','China','Thailand','Japan','Philippines']
australia = ['New Zealand','Australia']

for (row, col) in mh.iterrows():

    if str(col.Country) in north_america:
        mh_subset['Country'].replace(to_replace=col.Country, value='North_America', inplace=True)

    if str(col.Country) in south_america:
        mh_subset['Country'].replace(to_replace=col.Country, value='South_America', inplace=True)

    if str(col.Country) in europe:
        mh_subset['Country'].replace(to_replace=col.Country, value='Europe', inplace=True)

    if str(col.Country) in africa:
        mh_subset['Country'].replace(to_replace=col.Country, value='Africa', inplace=True)

    if str(col.Country) in asia:
        mh_subset['Country'].replace(to_replace=col.Country, value='Asia', inplace=True)

    if str(col.Country) in australia:
        mh_subset['Country'].replace(to_replace=col.Country, value='Australia', inplace=True)

print(mh_subset['Country'].unique())
mh_subset.head(10)



```
